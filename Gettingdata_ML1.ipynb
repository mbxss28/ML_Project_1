```python
#Used to assess the ChEMBL database
from chembl_webresource_client.new_client import new_client
#To import data into dataframe
import pandas as pd
```

```python
# Searching database for chemicals associated with Alzheimer's
target = new_client.target
target_query = target.search('Alzheimer')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
#Selecting which molecule will be focused on
selected_target = targets.target_chembl_id[2]
#The selected target here is the Gamma secretase single protien protein 
selected_target
```

```python
activity = new_client.activity
#IC50 is the half maximum inhibitory concentration
#In different biological datasets, this could be a different value, such as EC50 
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
#Results stored in a dictionary
res
```

```python
#converts dictionary data structure to pandas dataframe
df = pd.DataFrame.from_dict(res)
```

```python
df
#Type should all be IC50
#The lower the number for value, the better, since it requires less to inhibit a protein.
```

```python
#Creates CSV file from dataframe 
df.to_csv('gamma_raw.csv', index=False)
```
